```python
import requests
```

```python
baseurl = 'https://en.wikipedia.org'
```

```python
page = requests.get(baseurl + "/wiki/List_of_leaf_vegetables")
page
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
veg_list = soup.select('table.wikitable tr')
```

```python
first_col = [x.find('td').find('a') for x in veg_list[1:]]
```

```python
names = [x.find_all('td')[1].text for x in veg_list[1:]]
```

```python
url_txt = [ (baseurl + x['href'], x.text) for x in first_col]
```

```python
name_map=dict()
```

```python
for (idx, (url, sciname)) in enumerate(url_txt):
    name_map[sciname] = names[idx]
```

```python
url_map = dict()
```

```python
for (url, name) in url_txt:
    url_map[name] = requests.get(url)
```

```python
soup_map = dict()

for (name, itempage) in url_map.items():
    if (itempage.status_code == 200):
        isoup = BeautifulSoup(itempage.content, 'html.parser')
        soup_map[name] = isoup
```

```python
url_map['Xanthosoma brasiliense'].url
```

```python
isoup = soup_map['Xanthosoma brasiliense']
```

```python
def select_img(img_list):
    for x in img_list:
        alt = x['alt']
        if alt != None and alt != '' and (alt.lower().endswith("jpg")) or (alt.lower().endswith("jpeg")):
            return x
        
    return img_list[0]
    return None
```

```python
isoup.select('a img')
```

```python
select_img(isoup.select('a img'))
```

```python
from IPython.core.display import display, HTML
```

```python
display(HTML(str(select_img(isoup.select('a img')))))
```

```python
fullHtml="<table><tbody><tr><th>Scientific Name</th><th>Name</th><th>Image</th></tr>";
for (name, isoup) in soup_map.items():
    imHtml = str(select_img(isoup.select('a img')))
    rowHtml = "<tr><td>" + name + "</td><td>" + name_map[name] + "</td><td>" + imHtml + "</td></tr>"
    fullHtml = fullHtml + rowHtml
```

```python
display(HTML(fullHtml))
```
